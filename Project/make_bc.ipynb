```python
import numpy as np
import csv
# change the file name for your experiment
with open('simulation.getting_colder.csv', 'w') as file: 
    writer = csv.writer(file)
    writer.writerow(['Day of Simulation', 'T_atm_pole', 'T_atm_eq'])

    for day in range(1, 1001):
        T_atm_pole = -(5/365) * day + 0 # you can change this function
        T_atm_equator = 15 # you can change this function

        writer.writerow([day, T_atm_pole, T_atm_equator])
```

```python
import numpy as np
import csv

# change the file name for your experiment
with open('simulation.seasons.csv', 'w') as file: 
    writer = csv.writer(file)
    writer.writerow(['Day of Simulation', 'T_atm_pole', 'T_atm_eq'])

    for day in range(1, 1001):
        T_atm_equator = 5 * np.sin(2*np.pi*(day-1)/365) + 26 # you can change this function
        T_atm_pole = 10 * np.sin(2*np.pi*(day-1)/365) + 6 # you can change this function

        writer.writerow([day, T_atm_pole, T_atm_equator])
```

```python
import numpy as np
import csv

# change the file name for your experiment
with open('simulation.ice_age.csv', 'w') as file: 
    writer = csv.writer(file)
    writer.writerow(['Day of Simulation', 'T_atm_pole', 'T_atm_eq'])

    for day in range(1, 20*365+1):
        T_atm_equator = 5 * np.sin(2*np.pi*(day-1)/365) + 10 # you can change this function
        T_atm_pole = (10 * np.sin(2*np.pi*(day-1)/365) - 5) + (day * (5/(20*365))) # you can change this function

        writer.writerow([day, T_atm_pole, T_atm_equator])
```
