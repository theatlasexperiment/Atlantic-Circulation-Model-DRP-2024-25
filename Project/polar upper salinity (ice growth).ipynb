```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#To do: calculate change in salinity of polar upper box as a function of change in ice thickness
#ice thickness is a function of temperature change 
```

```python
#constants:

k_ice = 2.2 #thermal conductivity of ice in W/mÂ·K
latentheat =  334000 #latent heat of fusion of ice in J/kg
rho_ice = 916 #density of ice in kg/m^3 at 0 degrees C
ice_S = 5 #salinity of ice in psu

#note: thermal conductivity actually depends on temperature; latent heat on salinity
```

```python
#timestep:
a = 0 
total_time = 10000 #arbitrary
stepsize = 1 
time = np.arange(0,total_time, stepsize)

Tatm = 15*np.sin(- np.pi * time / total_time) #arbitrary sinusoid wave for atmospheric temperature
Tmelt = 0 #in Celsius 
D = 100 #box depth in meters, arbitrary

#note: melting point of sea ice is not in fact 0 degrees C, and depends on salinity 

plt.plot(Tatm)
```

```python
h_init = 0.1 #initial ice thickness
Sw_init = 35 #initial seawater salinity in psu
```

```python
h_total = h_init
Sw_total = Sw_init
const = 2*k_ice/(latentheat*rho_ice)

ice_thickness = [] 
Sw_salinity = []
waterdepthbox = []

for t in time:
    G = np.sqrt(const * (Tmelt - Tatm[t])*stepsize)  #G is change in ice thickness 
    #to resolve: negative values in sqrt with ice melting 
    
    h_total += G
    
    waterdepth = D - h_total
    
    d_Sw = ice_S * G * stepsize /waterdepth #d_Sw is change in seawater salinity 
    
    Sw_total += d_Sw
    
    ice_thickness.append(h_total)
    Sw_salinity.append(Sw_total)
    waterdepthbox.append(waterdepth)
```

```python
plt.plot(time, ice_thickness)
```

```python
plt.plot(time, Sw_salinity)
```

```python
plt.plot(time, waterdepthbox)
```
